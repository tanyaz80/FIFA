```python
def clean_Hits(x):
    if type(x) is str:
        if x.find('K'):
            s = x.replace('K', '000')
            s=  x.replace('.', '')
            
        return float(s[0])
        
    else:
        return float(x)
  
df_test=data2['Hits'].apply(clean_Hits)
print(df_test)
```

```python
#Height: 
def clean_height(x):
    if type(x) is str:
        s = x.replace('"', '')
        s=  int(s.split("'")[0])*30.48 + int(s.split("'")[1])*2.54
        return s
        
    else:
        return x
  
df_test=data2['Height'].apply(clean_height)
print(df_test)
```

```python
def clean_stars(x):
    if type(x) is str:
        if x.find(' ★') or x.find ('★'):
            s = x.replace(' ★', '') 
            s = x.replace('★', '')
           
            s = float(s)
        return s
        
    else:
        return x
df_test=data2['IR'].apply(clean_stars)
print(df_test)
```

```python
# Function fix plis minus problem

def clean_plus_minus(x):
    if type(x) is str:
        if x.find('+-') or x.find('+'):
            s = x.split('+')
            return float(s[0])
        elif x.find('-'):
            s = x.split('-')
            return float(s[0])
    else:
        return x

df_test=data2['CDM'].apply(clean_plus_minus)

print(df_test)
```

```python
def clean_wage(x):
    if type(x) is str:

        s = x.replace('€', '').replace('K', '000')

        return float(s)
        
    else:
        return x
  
df_test=data2['Wage'].apply(clean_wage)
print(df_test)
```

```python
#Weight fixed

def clean_weight(x):
    if type(x) is str:
        if x.find('lbs'):
            s = x.replace('lbs', '')
            s = float(s)*  0.45359237
        return s
        
    else:
        return x
  
df_test=data2['Weight'].apply(clean_weight)
print(df_test)

```

```python
def clean_value(x):
    if type(x) is str:
        s = x.replace('€', '').replace('K', "000").replace('M', '000000').translate({ord('.'): None})
        
        return float(s)
        
    else:
        return x
  
df_test=data2['Release Clause'].apply(clean_value)

print(df_test)
```
